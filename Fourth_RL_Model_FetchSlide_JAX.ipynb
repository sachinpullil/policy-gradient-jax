#Code

##Building The Agent

###Neural Network

```python
def MLP(layers, activation=jnn.relu, final_activation=jnp.tanh):
  def init(rng_key):
      def init_layer(key, d_in, d_out):
          k1, k2 = jax.random.split(key)
          glorot_stddev = 1. / jnp.sqrt((d_in + d_out) / 2.)
          W = glorot_stddev * jax.random.normal(k1, (d_in, d_out))
          b = jnp.zeros(d_out)
          return W, b
      key, *keys = jax.random.split(rng_key, len(layers))
      params = list(map(init_layer, keys, layers[:-1], layers[1:]))
      return params
  def apply(params, inputs):
      for W, b in params[:-1]:
          outputs = jnp.dot(inputs, W) + b
          inputs = activation(outputs)
      W, b = params[-1]
      outputs = jnp.dot(inputs, W) + b
      final_outputs = final_activation(outputs)
      return final_outputs
  return init, apply
```

###Optimizer

```python
#Jax code
@optimizer
def adam(step_size, b1=0.9, b2=0.999, eps=1e-8):
    step_size = make_schedule(step_size)
    def init(x0):
        m0 = jnp.zeros_like(x0)
        v0 = jnp.zeros_like(x0)
        return x0, m0, v0
    def update(i, g, state):
        x, m, v = state
        m = (1 - b1) * g + b1 * m  # First  moment estimate.
        v = (1 - b2) * jnp.square(g) + b2 * v  # Second moment estimate.
        mhat = m / (1 - jnp.asarray(b1, m.dtype) ** (i + 1))  # Bias correction.
        vhat = v / (1 - jnp.asarray(b2, m.dtype) ** (i + 1))
        x = x - step_size(i) * mhat / (jnp.sqrt(vhat) + eps)
        return x, m, v
    def get_params(state):
        x, _, _ = state
        return x
    return init, update, get_params
```

###Deep Learning Model/Agent

```python
class DRLModel: 
    # Initialize the class
    def __init__(self, layers, learning_rate, rng_key=jax.random.PRNGKey(0)):
        # MLP init and apply functions
        self.net_init, self.net_apply = MLP(layers)
        params = self.net_init(rng_key)
        self.learning_rate = learning_rate

        # Optimizer initialization and update functions
        lr = exponential_decay(self.learning_rate, decay_steps=1000, decay_rate=0.999)
        self.opt_init, \
        self.opt_update, \
        self.get_params = adam(lr)
        self.opt_state = self.opt_init(params)

    def set_initial_params(self, params):
        self.opt_state = self.opt_init(params)

    def loss(self, params, batch):
        inputs, targets = batch
        outputs = self.net_apply(params, inputs)
        loss = jnp.mean((targets - outputs)**2)
        return loss

    @partial(jit, static_argnums=(0,))
    def step(self, i, opt_state, batch):
        params = self.get_params(opt_state)
        gradients = grad(self.loss)(params, batch)
        return self.opt_update(i, gradients, opt_state)

    def train(self, training_data, iter_no):
        self.opt_state = self.step(iter_no, self.opt_state, training_data)  

    @partial(jit, static_argnums=(0,))
    def predict(self, params, inputs):
        outputs = self.net_apply(params, inputs)
        return outputs

    def getloss(self, training_data):
        params = self.get_params(self.opt_state)
        training_loss = self.loss(params, training_data)
        return training_loss
```

```python
env = gym.make("FetchSlide-v1").env
obs_size = env.observation_space["desired_goal"].shape[0] + \
              env.observation_space["observation"].shape[0]
action_size = env.action_space.shape[0]
HIDDEN_SIZE = 256

layers = [obs_size, HIDDEN_SIZE, HIDDEN_SIZE, HIDDEN_SIZE, action_size]
agent = DRLModel(layers, learning_rate=0.010)
```

###Select an Action

```python
def select_continuous_action(PRNGkey, state, params, noise_std=0.7):
  state_jnp = jnp.concatenate([state['desired_goal'], state['observation']])
  action_means = agent.predict(params, state_jnp)
  action = jax.random.multivariate_normal(PRNGkey, mean=action_means.flatten(), cov=noise_std*jnp.eye(action_size))
  action_clipped = jnp.clip(action, env.action_space.low[0], env.action_space.high[0])
  return np.asarray(action_clipped)

#Consider discretizing the action space.
```

##Training

```python
data = np.load('fourth_model_400_iter.npy')

running_index = 0
loaded_params = []
for i in range(len(layers)-1):
  d1=layers[i]
  d2=layers[i+1]
  W_flat = data[running_index:running_index+d1*d2]
  W = jnp.asarray(W_flat.reshape((d1,d2)))
  running_index += d1*d2
  b = jnp.asarray(data[running_index:running_index+d2])
  running_index +=d2
  loaded_params.append((W,b))

agent.set_initial_params(loaded_params)
```

```python
BATCH_SIZE = 100
PERCENTILE = 70
REWARD_GOAL = -80

from collections import namedtuple

Episode = namedtuple('Episode', field_names=['reward', 'steps'])
EpisodeStep = namedtuple('EpisodeStep', field_names=['observation', 'action'])
```

```python
iter_no = 0
step_count = 0
reset_count = 0
full_batch = []
full_batch_steps = []
full_states = []
full_acts = []
batch = []
episode_steps = []
episode_reward = 0.0
EPISODE_MAX_STEPS = 200
MAX_ITERATIONS = 100
reward_mean = -1*EPISODE_MAX_STEPS
reward_mean_collec = np.full((10,), reward_mean)

state = env.reset(seed=reset_count)

rng = np.random.default_rng()
true_random = rng.integers(10000)

initial_key = jax.random.PRNGKey(true_random)
step_keys = jax.random.split(initial_key, EPISODE_MAX_STEPS)
curr_params = agent.get_params(agent.opt_state)

goal = np.zeros(3)
puck_initial = np.zeros(3)
puck_later = np.zeros(3)
initial_dist = 0.0
later_dist = 0.0

print("Loading model after 300 iterations..")

while reward_mean < REWARD_GOAL and iter_no < MAX_ITERATIONS:
  # action = []
  # if iter_no < 50:
  #   action = env.action_space.sample()
  # else:
  action = select_continuous_action(step_keys[step_count], state, curr_params) #Get an action for current state
  next_state, reward, dummy, info  = env.step(action)                           #Get results of the action
  episode_is_done = info["is_success"]
  step_count +=1

  state_np = np.concatenate([state['desired_goal'], state['observation']])
  episode_steps.append(EpisodeStep(observation=state_np, action=action))        #Add it to our list of steps for this episode
  episode_reward += reward                                                      #Accumulate reward

  # #At 50 steps, check if puck is closer. Cut episode if not. -- DONE
  # if 1 == step_count:
  #   puck_initial[0] = state['observation'][3]
  #   puck_initial[1] = state['observation'][4]
  #   puck_initial[2] = state['observation'][5]
  #   goal[0] = state['desired_goal'][0]
  #   goal[1] = state['desired_goal'][1]
  #   goal[2] = state['desired_goal'][2]
  #   initial_dist = np.linalg.norm(goal-puck_initial)
  # elif 50 == step_count:
  #   puck_later[0] = state['observation'][3]
  #   puck_later[1] = state['observation'][4]
  #   puck_later[2] = state['observation'][5]
  #   later_dist = np.linalg.norm(goal-puck_later)
  #   if not initial_dist > later_dist:
  #     #Throw away this episode because puck hasn't moved or has moved away from goal
  #     next_state = env.reset(seed=reset_count)                                    #Reset
  #     episode_steps = []                                                          #Reset
  #     episode_reward = 0.0                                                        #Reset
  #     step_count = 0
  #     state = next_state



  if episode_is_done or EPISODE_MAX_STEPS==step_count:                          #Episode finished because goal or max steps
    batch.append(Episode(reward=episode_reward, steps = episode_steps))         #Create the episode using the steps and final reward
    reset_count +=1
    next_state = env.reset(seed=reset_count)                                    #Reset
    episode_steps = []                                                          #Reset
    episode_reward = 0.0                                                        #Reset
    step_count = 0
    initial_key = jax.random.PRNGKey(true_random+reset_count)
    step_keys = jax.random.split(initial_key, EPISODE_MAX_STEPS)

    if(len(batch) == BATCH_SIZE):                                               #New set of batches ready --> select best ones
      reward_list = list(map(lambda s: s.reward, batch))                        #Make list of all rewards
      reward_mean = float(np.mean(reward_list))                                 #Find mean of rewards
      elite_candidates = batch                                                  #Candidates for elitism
      #returnG = list(map(lambda s: s.reward * (GAMMA ** len(s.steps)), 
      #                   elite_candidates))  
      #The above only works when reward is positive.                                    #Make list of DISCOUNTED reward for all elite candidates
      reward_bound = np.percentile(reward_list, PERCENTILE)                     #Find the top 30 percentile within this list. Though code seems to do top 70 percentile
                                                                                #The bound will tell us the cutoff score to check if an episode is part of the elite
      train_obs = []                                                            #Observations for training
      train_act = []                                                            #Actions for training
      elite_batch = []                                                          #Batch of elites

      for example, example_reward in zip(elite_candidates, reward_list):        #Zip makes a tuple out of each element in elite_candidates with each element in returnG
                                                                                #For each example episode and discounted reward for that episode..
        if example_reward > reward_bound:                                        #If episode is part of the elite
          train_obs.extend(map(lambda step: step.observation, example.steps))   #Add all observations for steps within episode to training list
          train_act.extend(map(lambda step: step.action, example.steps))        #Add all actions for steps within episode to training list
          elite_batch.append(example)                                           #Add the episode itself to the elite
          full_batch_steps.append(len(example.steps))
      
      full_batch.extend(elite_batch)                                                  #Our full batch is our updated elite batch
      full_states.extend(train_obs)                                                         #List of training observations
      full_acts.extend(train_act)                                                          #List of training actions
      print("iter_no = {}, elite_batch size = {}, full_batch size = {}, reward_mean = {}, reward_bound = {}".format(iter_no, len(elite_batch), len(full_batch), reward_mean, reward_bound))

      #If full_batch grows too big, we delete first twenty episodes
      if 100<len(full_batch):
        steps_length = np.sum(full_batch_steps[0:20])
        del full_batch[0:20]
        del full_batch_steps[0:20]
        del full_states[0:steps_length]
        del full_acts[0:steps_length]
        print("Deleted 20 episodes and {} steps/acts".format(steps_length))

      if len(full_batch) != 0:                                                  # just in case empty during an iteration.
        training_data = (jnp.asarray(full_states), jnp.asarray(full_acts))
        agent.train(training_data, iter_no)
        reward_mean_collec[iter_no%10] = reward_mean
        if iter_no%10 == 0:
          loss = agent.getloss(training_data)
          print("%d: loss=%.3f, average reward_mean=%.3f" % (iter_no, loss.item(), 
                                                   np.mean(reward_mean_collec)))
        iter_no += 1                                                            #Notice how iter_no refers to number of BATCHES that are used to train
        curr_params = agent.get_params(agent.opt_state)

      batch = []                                                                #Reset batch list
  
  state = next_state                                                            #Get ready for next loop
```

##Saving

```python
opt_params = agent.get_params(agent.opt_state)

param_count = np.sum(np.asarray([layers[i]*layers[i+1] + layers[i+1] for i in range(len(layers)-1)]))

params_np = np.zeros((param_count,))
running_index = 0
for W, b in opt_params:
  W_flat = np.asarray(W.flatten())
  W_length = W_flat.shape[0]
  b_flat = np.asarray(b.flatten())
  b_length = b_flat.shape[0]

  params_np[running_index:running_index+W_length] = W_flat
  running_index += W_length
  params_np[running_index:running_index+b_length] = b_flat
  running_index += b_length

np.save('fourth_model_400_iter.npy', params_np)
```
